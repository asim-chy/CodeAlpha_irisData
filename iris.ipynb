```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
print("=",60)
print("Total Dataset Information")
print("=",60)
print("Total Samples",X.shape[0])
print("Total Features",X.shape[1])
print("Total Classes",len(np.unique(y)))
```

```python
k=5
kf=KFold(
    n_splits=k,
    shuffle=True,
    random_state=42
)
model = KNeighborsClassifier(n_neighbors=3)
accuracy_list =[]

print("\n")
for fold_idx, (train_index, test_index) in enumerate(kf.split(X)):
    print("=" * 70)
    print(f"Fold {fold_idx + 1}")
    print("-" * 70)

```

```python
print("Training Samples : ",len(train_index))
print("Testing Samples : ",len(test_index))

print("\nTraining Index")
print(train_index)

print("\nTesting Index")
print(test_index)

x_train, x_test = X[train_index], X[test_index]
y_train, y_test = y[train_index], y[test_index]

model.fit(x_train,y_train)
prediction = model.predict(X=x_test)
accuracy = accuracy_score(y_true=y_test, y_pred=prediction)
accuracy_list.append(accuracy)
print("\nAccuracy = %.2f%%"%(accuracy*100))

print("\n")
print("="*60)
print("FINAL RESULT")
print("="*60)
print(f"Accuracy Score : {np.mean(accuracy_list)*100}")
```
